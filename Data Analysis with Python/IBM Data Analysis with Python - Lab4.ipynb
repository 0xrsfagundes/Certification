```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('csv/automobileEDA.csv')
df = df.sort_index(axis=1)

df.info()
```

# Linear Regression

<p>Simple Linear Regression: help us understand the relationship between two variables</p>

Yhat = a + bX

X: Predictor Variable
Y: Response Variable

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm
```

```python
X = df[['highway-mpg']] #Linear regression models expect input data to be in a two-dimensional format
Y = df['price']

lm.fit(X,Y)
```

```python
Yhat=lm.predict(X)
Yhat[0:5] 
```

```python
lm.intercept_
```

```python
lm.coef_
```

### Yhat = a + bX

### Yhat = lm.intercept_ + lm.coef_ * X

### Price = 38423.31 - 821.73 * highway-mpg

# Polynomial Regression and Pipelines

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
x = df['highway-mpg']
y = df['price']
```

```python
# Here we use a polynomial of the 3rd order (cubic) 
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

```python
PlotPolly(p, x, y, 'highway-mpg')
```

```python
np.polyfit(x, y, 3)
```

```python
f1 = np.polyfit(x, y, 11)
p1 = np.poly1d(f1)
print(p1)
PlotPolly(p1,x,y, 'Highway MPG')
```

# Pipeline
Data Pipelines simplify the steps of processing the data

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model',LinearRegression())]

pipe=Pipeline(Input)
pipe
```

# Measures for In-Sample Evaluation

higher R-squared value is a better fit<br/>
smallest MSE value is a better fit for the data.

## Simple Linear Regression

```python
#highway_mpg_fit
lm.fit(X, Y)
# Find the R^2
print('The R-square is: ', lm.score(X, Y))
```

```python
Yhat=lm.predict(X)
print('The output of the first four predicted value is: ', Yhat[0:4])
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(df['price'], Yhat)
print('The mean square error of price and predicted value is: ', mse)
```

## Multiple Linear Regression

```python
# fit the model 
lm.fit(Z, df['price'])
# Find the R^2
print('The R-square is: ', lm.score(Z, df['price']))
```

```python
Y_predict_multifit = lm.predict(Z)
```

```python
print('The mean square error of price and predicted value using multifit is: ', \
      mean_squared_error(df['price'], Y_predict_multifit))
```

## Polynomial Fit

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y, p(x))
print('The R-square value is: ', r_squared)
```

```python
mean_squared_error(df['price'], p(x))
```
