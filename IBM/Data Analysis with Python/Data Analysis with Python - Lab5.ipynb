```python
import pandas as pd

df = pd.read_csv('csv/automobileEDA.csv')

df.sort_index(axis=1, inplace = True)

df.head()
```

```python
df.info()
```

```python
df_corr = df.corr(numeric_only=True)

df_corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df_corr, annot=True, annot_kws={"fontsize":5})
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df_corr.loc[['bore','stroke','compression-ratio','horsepower']][['bore','stroke','compression-ratio','horsepower']], annot=True, annot_kws={"fontsize":6})
plt.show()
```

```python
# Engine size as potential predictor variable of price
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

```python
df[['engine-size','price']].corr()
```
