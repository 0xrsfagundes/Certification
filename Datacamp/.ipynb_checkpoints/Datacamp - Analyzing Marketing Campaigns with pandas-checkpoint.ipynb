# Datacamp - Analyzing Marketing Campaigns with pandas

## Chapter 1

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('csv/marketing.csv',parse_dates=['date_canceled', 'date_served', 'date_subscribed'])

df.sort_index(axis=1, inplace=True)

df.head()
```

```python
df.info()
```

```python
df.describe(include='all', datetime_is_numeric=True)
```

```python
# Mapping for channels
channel_dict = {"House Ads": 1, "Instagram": 2, 
                "Facebook": 3, "Email": 4, "Push": 5}

# Map the channel to a channel code
df['channel_code'] = df['subscribing_channel'].map(channel_dict)

df.head()
```

```python
# Add a DoW column
df['DoW'] = df['date_subscribed'].dt.dayofweek

df.head()
```

Visualizing daily marketing reach

```python
# Group by date_served and count number of unique user_id's
# nunique = returns the count of unique values in the specified axis, which could be a column or a row
daily_users = df.groupby(['date_served'])['user_id'].nunique() 

# Print head of daily_users
daily_users.head()
```

```python
# Plot daily_subscribers
daily_users.plot()

# Include a title and y-axis label
plt.title('Daily users')
plt.xlabel('Date')
plt.ylabel('Number of users')

# Rotate the x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

## Chapter 2

Calculating conversion rate

```python
total = df['user_id'].nunique()

subscribers = df[df['converted']== True]['user_id'].nunique()

CR = subscribers/total

print('\nTotal:', total,'\nSubscribers:',subscribers,'\nCR:',CR)
```

```python
daily_subscribers = df[df['converted']== True].groupby(['date_served'])['user_id'].nunique()

daily_subscribers.head()
```

```python
merged_df = pd.merge(daily_users,daily_subscribers,on='date_served')

merged_df = merged_df.rename(columns={'user_id_x':'users','user_id_y':'subscribers'})

merged_df['CR'] = merged_df['subscribers']/merged_df['users']

merged_df = merged_df.reset_index()

merged_df.head()
```

```python
merged_df.plot(x='date_served',y=['subscribers'])
plt.show()
```

```python
merged_df.plot(x='date_served',y=['CR'])
plt.show()
```

Calculating retention rate


```python
daily_retained = df[df['is_retained']==True].groupby(['date_served'])['user_id'].nunique()

daily_retained.head()
```

```python
merged_df2 = pd.merge(merged_df,daily_retained, on='date_served')

merged_df2.rename(columns={'user_id':'retained'},inplace=True)

merged_df2['RR'] = merged_df2['retained']/merged_df2['users']

merged_df2.head()
```

English speakers

```python
languages_df = df.groupby(['date_served','language_displayed'])['user_id'].nunique()

newLanguages_df = languages_df.to_frame()

newLanguages_df.reset_index(inplace=True)

newLanguages_df.head()
```

```python
newLanguagesPivot_df = newLanguages_df.pivot(index='date_served',columns='language_displayed', values='user_id')
newLanguagesPivot_df.reset_index(inplace=True)
newLanguagesPivot_df.head()
```

```python
merged_df3= pd.merge(merged_df2,newLanguagesPivot_df,on='date_served')

merged_df3.fillna(0,inplace=True)

merged_df3.head()
```

```python
merged_df3['CR_Arabic'] = merged_df3['Arabic'] / merged_df3['users']
merged_df3['CR_English'] = merged_df3['English'] / merged_df3['users']
merged_df3['CR_German'] = merged_df3['German'] / merged_df3['users']
merged_df3['CR_Spanish'] = merged_df3['Spanish'] / merged_df3['users']

merged_df3.head()
```

```python
groupTotalUsers = df.groupby('language_displayed')['user_id'].nunique()

groupSubscribers = df[df['converted']==True].groupby('language_displayed')['user_id'].nunique()

languageConversionRate = groupSubscribers/groupTotalUsers

languageConversionRate.plot(kind='bar')
plt.show()
```

```python
merged_df3[['CR_Arabic','CR_English','CR_German','CR_Spanish']].plot()
plt.show()
```

Marketing channels across age groups


```python
channel_age = df.groupby(['marketing_channel','age_group'])['user_id'].count().to_frame()

channel_age
```

Unstack method

```python
channel_age_df1 = pd.DataFrame(channel_age.unstack(level = 1))

channel_age_df1
```

Pivot method

```python
channel_age.reset_index(inplace=True)
channel_age_df2 = channel_age.pivot(index='marketing_channel',columns='age_group',values='user_id')
channel_age_df2
```

```python
ax = channel_age_df2.plot(kind='bar')

for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=6, rotation=90)
    
plt.title('Marketing channels by age group')
plt.xlabel('Age group')
plt.ylabel('Users')
plt.show()
```

Grouping and counting by multiple columns


```python
retention_total = df.groupby(['date_subscribed','subscribing_channel'])['user_id'].nunique().to_frame()

retention_subs = df[df['is_retained']==True].groupby(['date_subscribed','subscribing_channel'])['user_id'].nunique().to_frame()

retention_rate = retention_subs/retention_total

retention_rate.reset_index(inplace=True)

retention_rate_pivot = retention_rate.pivot(index='date_subscribed', columns='subscribing_channel',values='user_id')

retention_rate_pivot.fillna(0, inplace=True)

retention_rate_pivot.head()
```

```python
retention_rate_pivot.plot()
plt.title('Retention Rate by Subscribing Channel')
plt.xlabel('Date Subscribed')
plt.ylabel('Retention Rate (%)')
plt.show()
```

Building a conversion function


```python
def conversion_rate(dataframe,column_names)
    df_total = dataframe.groupby('converted')['user_id'].nunique()

    df_conv = dataframe[dataframe['converted']==True].groupby(column_names)['user_id'].nunique()

    conversion_rate = df_conv/df_total

    return conversion_rate
```
