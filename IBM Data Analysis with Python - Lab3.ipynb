# Corr - calculate the correlation between columns of a DataFrame, using as defatult Pearson correlation coefficient

The coefficient ranges from -1 to 1:

 1: indicates a perfectly positive correlation
 0: indicates no correlation
-1: indicates a perfectly negative correlation

```python
df_corr = df.corr(numeric_only=True)

df_corr
```


# Heatmap

annot=True - adds the correlation coefficients to the heatmap
cmap='coolwarm' - sets the color scheme for the heatmap

annot_kws: dictionary of keyword arguments
    fontsize: Sets the font size of the annotations (e.g. {"fontsize": 12}).
    fontweight: Sets the font weight of the annotations (e.g. {"fontweight": "bold"}).
    color: Sets the color of the annotations (e.g. {"color": "red"}).
    ha: Sets the horizontal alignment of the annotations (e.g. {"ha": "center"}).
    va: Sets the vertical alignment of the annotations (e.g. {"va": "center"}).
    rotation: Sets the rotation angle of the annotations (e.g. {"rotation": 90}).
    bbox: Sets the bounding box properties of the annotations (e.g. {"bbox": {"facecolor": "white", "alpha": 0.5, "edgecolor": "red"}}).

```python
sns.heatmap(df_corr, annot=True, annot_kws={"fontsize":5}, cmap='coolwarm')
plt.show()
```

```python
# plt.subplots() - creates a figure and a set of subplots, and returns a tuple containing both.
# ax.pcolor() - creates a pseudocolor plot of a 2D array or a DataFrame, which can be used to create a heatmap plot.
# cmap - specifies the colormap to use for the plot. 
# row_labels and col_labels - store the labels for the rows and columns of the heatmap plot.
# ax.set_xticks() and ax.set_yticks() - are used to set the positions of the ticks on the x and y axes.
# ax.set_xticklabels() and ax.set_yticklabels() - are used to set the labels for the ticks on the x and y axes.
# plt.xticks() - is used to rotate the x-axis tick labels by 90 degrees if they are too long.
# fig.colorbar() - adds a colorbar to the plot created by ax.pcolor()
# plt.show() - display plot

fig, ax = plt.subplots()
im = ax.pcolor(df_grouped_pivot, cmap='Reds')

#label names
row_labels = df_grouped_pivot.columns.levels[1]
col_labels = df_grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

plt.title('Relationship between Car Body Style and Drive Wheel with Sales')

# Add a secondary y-axis on the right-hand side of the plot
ax2 = ax.twinx()

# Set the label for the secondary y-axis
ax2.set_ylabel('Price')

# Move the label a little bit to the right
ax2.yaxis.set_label_coords(1.3, 0.5)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
