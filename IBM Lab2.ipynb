Data Analysis with Python

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df = pd.read_csv('auto.csv', names=headers)

df.head()
```


# Replace '?' to NAN


```python
df.replace('?', np.nan, inplace=True)
df.sort_index(axis=1, inplace = True)

df.head()
```

```python
df.info()
```

```python
nan_columns = []

for column in df.columns:
    if df[column].isna().sum() > 0:
        nan_columns.append([column,df[column].isna().sum()])
        
df_nan_columns = pd.DataFrame(nan_columns, columns =['column','nan_count'])

df_nan_columns
```

```python
df[df_nan_columns['column']].head()
```


# Replace NAN to mean values


```python
#axis=0 -> column
df['bore'].replace(np.nan, df['bore'].astype('float').mean(axis=0), inplace=True)
df['horsepower'].replace(np.nan, df['horsepower'].astype('float').mean(axis=0), inplace=True)
df['normalized-losses'].replace(np.nan, df['normalized-losses'].astype('float').mean(axis=0), inplace=True)
df['peak-rpm'].replace(np.nan, df['peak-rpm'].astype('float').mean(axis=0), inplace=True)
df['price'].replace(np.nan, df['price'].astype('float').mean(axis=0), inplace=True)
df['stroke'].replace(np.nan, df['stroke'].astype('float').mean(axis=0), inplace=True)


df[df_nan_columns['column']].head()
```

```python
df['num-of-doors'].value_counts(dropna=False)
```


# Replace NAN to Max values


```python
#idmax = returns the index of the max value in a series
df['num-of-doors'].value_counts(dropna=False).idxmax()
```

```python
df['num-of-doors'].replace(np.nan, df['num-of-doors'].value_counts().idxmax(), inplace= True)
```

```python
df['price'].value_counts(dropna=False)
```

```python
df[df['price'].isna()]
```


# Drop NAN Price rows


```python
#axis=0 -> Rows
df.dropna(subset='price', axis=0, inplace =True)
```

```python
#drop=True, it will drop the old index column and only keep the new index column in the DataFrame
df.reset_index(drop=True, inplace=True)

df.head()
```


# Fix columns datatypes


```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```


# Data Standardization


```python
df['city-L/100km'] = 235/df['city-mpg']

df[['city-L/100km','city-mpg']]
```


# Data Normalization

Normalization is the process of transforming values of several variables into a similar range

```python
df['length_norm'] = df['length']/df['length'].max()
df['width_norm'] = df['width']/df['width'].max()
df['height_norm'] = df['height']/df['height'].max()

df[['length_norm','length','width_norm','width', 'height_norm','height']]
```


# Binning / Categorization 


```python
df['horsepower'].head()
```

```python
df['horsepower'] = df['horsepower'].astype(int)

df['horsepower'].head()
```


# Histogram of bins


```python
plt.pyplot.hist(df["horsepower"])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(df["horsepower"].min(),df["horsepower"].max(),4)
```

cut() - create a new column in a DataFrame that categorizes the values into specified bins.

bins - The bins parameter specifies the intervals that values will be split into

labels parameter -  specifies the labels to assign to each bin

include_lowest=True parameter - ensures that the first interval includes the lowest value in the column

```python
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower"].value_counts()
```

```python
df["horsepower-binned"].value_counts()
```

```python
pyplot.bar(group_names, df["horsepower-binned"].value_counts())
```
