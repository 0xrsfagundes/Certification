```python
import pandas as pd

df = pd.read_csv('automobileEDA.csv')

df.sort_index(axis=1, inplace = True)

df.head()
```

```python
df.info()
```

```python
df_corr = df.corr(numeric_only=True)

df_corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df_corr[['bore','stroke','compression-ratio','horsepower']], annot=True, annot_kws={"fontsize":6})
plt.show()
```
